```python
import matplotlib.pyplot as plt
import numpy as np
```

## Tarefa 1 $\mapsto$ Gr√°ficos dos sinais

```python
data = np.loadtxt("output-signal1.dat")

plt.figure(figsize=(8, 8))
plt.plot(data[:,0], data[:,1], color='#b5b823', marker='o', linestyle='dashed', markersize=5, label = r'$a_1 = 2, a_2 = 4, \omega_1 = 4\pi(Hz), \omega_2 = 2.5\pi (Hz)$')

plt.xlim(0, 2.5 * np.pi)

plt.grid(True)
plt.ylabel("", fontsize=20)
plt.xlabel("", fontsize=20)
plt.legend(fontsize=12)

plt.savefig("sinal1.eps")
plt.show()
```
