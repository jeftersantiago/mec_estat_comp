```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def signals(path, label=""):
    data = np.loadtxt(path)
    plt.plot(data[:,0],data[:,1], color='black', linewidth = 1, label = label)
    plt.xlabel("Tempo")
    plt.ylabel("Amplitude")
    plt.legend()
    plt.show()
```

```python
signals("output-signal1.dat", "a = 2, b = 2, ...")
```

```python
original_signal = np.loadtxt('output-signal1.dat')
dft_inverse = np.loadtxt('output-inv-dft.dat')

plt.plot(original_signal[:,0], original_signal[:,1], color='red', label="Sinal original",  linewidth=1)

plt.plot(dft_inverse[:,0] ,dft_inverse[:,1], color='blue', label="Transformada inversa", 
         marker='o', linestyle='--', linewidth=0.5, markersize=3)
plt.xlabel("Tempo")
plt.ylabel("Amplitude do sinal")
plt.legend()
plt.show()
```
