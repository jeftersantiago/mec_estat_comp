```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
```

```python
item = "c"

evolucao = "saidas/evolucao-item-" + item + ".dat"
dft = "saidas/dft-item-" +  item + ".dat"

evolucao_img = "../relatorio/graficos/evolucao-item-" + item +".eps"
dft_img = "../relatorio/graficos/dft-item-" + item + ".eps"
```

```python
# Grfico simples de s 1 periodo completo
data = np.loadtxt(evolucao,   max_rows = 1000)

Y = data.T
X = np.linspace(0, 1, len(Y))

offset_increment = 1
offset = 0

plt.figure(figsize=(6, 6))
skip = 15
for i in range(0, Y.shape[1], skip):
   plt.plot(X, skip* Y[:, i] + skip * offset, color = "black")
   offset += offset_increment 

# print(offset)
# plt.yticks(y_ticks)
plt.xlabel("Espao (m)", fontsize=16)
plt.ylabel("Iterao", fontsize=16)
plt.xlim(0, 1)
plt.tight_layout()
plt.savefig(evolucao_img, dpi=300, bbox_inches='tight')
plt.show()
```

# Item A

```python
Y = np.loadtxt(dft)
freq = Y[:, 0]
val = Y[:, 1] #  * 1e-3

fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(freq, val, marker='.', s=200, color = "purple",  label = '$\\mathbb{R}^2 + \\mathbb{I}^2$')


plt.legend(fontsize=16)
plt.ylabel("Amplitude", fontsize=16)
plt.xlabel("Frequncia $(Hz)$", fontsize=16)

# ax.set_xticks([])
plt.legend(loc = "upper right", fontsize=16)

plt.xlim(0, 5000)

ax.xaxis.set_major_locator(MultipleLocator(150))

for i in range(0, 5000, 150):
    if i % 300 != 0:
        ax.axvline(x=i, color='darkorange', linestyle='--', linewidth=0.9)

for i in range(0, 5000, 150):
    if i % 300 == 0:
        ax.axvline(x=i, color='gray', linestyle='--', linewidth=0.9)

ax.set_xticklabels(ax.get_xticks(), rotation=30, ha='right')
ax.xaxis.set_major_formatter(FormatStrFormatter('%d'))
plt.savefig(dft_img, dpi=300, bbox_inches='tight')
plt.show()
```
