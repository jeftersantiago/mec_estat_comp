```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
```

# I

## Item (a)

```python
N = 1000
L = 1.0e0
r = 1.0e0
c = 300.0e0
dx = L/(N)
dt = r * dx / c
t = 0.03
nt = int(np.floor(t/dt))

print(dt)
print(nt)
```

```python
# Grfico simples de s 1 periodo completo
data = np.loadtxt("saida-tarefa1-a.dat", max_rows = 1000)

Y = data.T
X = np.linspace(0, 1, len(Y))

offset_increment = 1
offset = 0

plt.figure(figsize=(6, 6))
skip = 15
# Y = Y*dt
for i in range(0, Y.shape[1], skip):
   plt.plot(X, skip* Y[:, i] + skip * offset, color = "black")
   offset += offset_increment 

print(offset)
# plt.yticks(y_ticks)
plt.xlabel("Espao (m)", fontsize=16)
plt.ylabel("Iterao", fontsize=16)
plt.xlim(0, 1)
plt.tight_layout()

plt.savefig('../relatorio/graficos/graf-tarefa1-a.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Item (b)

```python
N = 9
data = np.loadtxt("saida-tarefa1-b.dat", max_rows =  2 * N)

Y = data.T
X = np.linspace(0, 1, len(Y))
fig = plt.figure(figsize=(6, 6))

ax = fig.add_subplot(111)
ax.set_xlabel('Espao (m)', fontsize = 16)
ax.set_ylabel('Iterao', fontsize = 16)

offset_increment = 1
offset = 0
for i in range(0, N+1+1+1+1+1+1+1): # Y.shape[1]):
    plt.plot(X, 5*(Y[:, i] + offset), color="black")
    offset += offset_increment 
#    print(i)

plt.plot(X, Y[:, 16] + 5*offset,marker=',', color = "red")

ax.set_yticks([])
plt.yticks()
plt.xlim(0, 1)

plt.savefig('../relatorio/graficos/graf-tarefa1-b.png', dpi=300, bbox_inches='tight')
plt.show()
```

# II

## Grfico da condio inicial

```python
X = np.linspace(0, 1, 100)
L = 1
Y = []
for x in X:
    if x < L/4:
        Y.append(x)
    else:
        Y.append((L-x)/3.0)
plt.figure(figsize=(10, 4))
plt.plot(X, Y, color="red", linewidth=2)

plt.xlabel("Espao (m)", fontsize=16)
plt.ylabel("Y0", fontsize = 16)

plt.grid(True, which='both', linestyle='', linewidth=0.5)
plt.tight_layout()

plt.ylim(0, 0.25)
plt.show()
```
